```python
def Fill_Hex(data, shape, height, radius, side_length, area, perimeter):
    # Set the component name in the metadata
    data['metadata']['component_name'] = shape

    # Calculate the coordinates for the hexagon vertices
    p1_x = round(side_length, 2)
    p1_y = 0

    p2_x = round(side_length * cos(300 * (math.pi) / 180), 2)
    p2_y = round(side_length * sin(300 * (math.pi) / 180), 2)

    p3_x = round(side_length * cos(240 * (math.pi) / 180), 2)
    p3_y = round(side_length * sin(240 * (math.pi) / 180), 2)

    p4_x = round(side_length * -1, 2)
    p4_y = 0

    p5_x = round(side_length * cos(120 * (math.pi) / 180), 2)
    p5_y = round(side_length * sin(120 * (math.pi) / 180), 2)

    p6_x = round(side_length * cos(60 * (math.pi) / 180), 2)
    p6_y = round(side_length * sin(60 * (math.pi) / 180), 2)

    # Calculate diameters and areas
    diameter = 2 * radius
    diameter_hex = 2 * side_length
    area_hex = (3 * math.sqrt(3) / 2) * (side_length ** 2)
    area_circle = math.pi * radius ** 2

    # External area and perimeters
    outer_area = area
    perimeter_hex = perimeter
    perimeter_circle = 2 * math.pi * radius

    # Set hexagon vertices in the data structure
    data['entities']['entity1']['points']['point1']['x'] = p1_x
    data['entities']['entity1']['points']['point1']['y'] = p1_y

    data['entities']['entity1']['points']['point2']['x'] = p2_x
    data['entities']['entity1']['points']['point2']['y'] = p2_y

    data['entities']['entity1']['points']['point3']['x'] = p3_x
    data['entities']['entity1']['points']['point3']['y'] = p3_y

    data['entities']['entity1']['points']['point4']['x'] = p4_x
    data['entities']['entity1']['points']['point4']['y'] = p4_y

    data['entities']['entity1']['points']['point5']['x'] = p5_x
    data['entities']['entity1']['points']['point5']['y'] = p5_y

    data['entities']['entity1']['points']['point6']['x'] = p6_x
    data['entities']['entity1']['points']['point6']['y'] = p6_y

    # Set the radius for the circle
    data['entities']['entity1']['curves']['circle']['radius'] = radius

    # Set the diameters
    data['entities']['entity1']['dimensions']['d_circle']['parameter']['value'] = diameter
    data['entities']['entity1']['dimensions']['d_hex']['parameter']['value'] = diameter_hex

    # Set hexagon profile points in the data structure
    data['entities']['entity1']['profiles']['profile1']['loops'][0]['profile_curves'][0]['start_point']['x'] = p6_x
    data['entities']['entity1']['profiles']['profile1']['loops'][0]['profile_curves'][0]['start_point']['y'] = p6_y

    data['entities']['entity1']['profiles']['profile1']['loops'][0]['profile_curves'][0]['end_point']['x'] = p1_x
    data['entities']['entity1']['profiles']['profile1']['loops'][0]['profile_curves'][0]['end_point']['y'] = p1_y

    data['entities']['entity1']['profiles']['profile1']['loops'][0]['profile_curves'][1]['start_point']['x'] = p1_x
    data['entities']['entity1']['profiles']['profile1']['loops'][0]['profile_curves'][1]['start_point']['y'] = p1_y

    data['entities']['entity1']['profiles']['profile1']['loops'][0]['profile_curves'][1]['end_point']['x'] = p2_x
    data['entities']['entity1']['profiles']['profile1']['loops'][0]['profile_curves'][1]['end_point']['y'] = p2_y

    data['entities']['entity1']['profiles']['profile1']['loops'][0]['profile_curves'][2]['start_point']['x'] = p2_x
    data['entities']['entity1']['profiles']['profile1']['loops'][0]['profile_curves'][2]['start_point']['y'] = p2_y

    data['entities']['entity1']['profiles']['profile1']['loops'][0]['profile_curves'][2]['end_point']['x'] = p3_x
    data['entities']['entity1']['profiles']['profile1']['loops'][0]['profile_curves'][2]['end_point']['y'] = p3_y

    data['entities']['entity1']['profiles']['profile1']['loops'][0]['profile_curves'][3]['start_point']['x'] = p3_x
    data['entities']['entity1']['profiles']['profile1']['loops'][0]['profile_curves'][3]['start_point']['y'] = p3_y

    data['entities']['entity1']['profiles']['profile1']['loops'][0]['profile_curves'][3]['end_point']['x'] = p4_x
    data['entities']['entity1']['profiles']['profile1']['loops'][0]['profile_curves'][3]['end_point']['y'] = p4_y

    data['entities']['entity1']['profiles']['profile1']['loops'][0]['profile_curves'][4]['start_point']['x'] = p4_x
    data['entities']['entity1']['profiles']['profile1']['loops'][0]['profile_curves'][4]['start_point']['y'] = p4_y

    data['entities']['entity1']['profiles']['profile1']['loops'][0]['profile_curves'][4]['end_point']['x'] = p5_x
    data['entities']['entity1']['profiles']['profile1']['loops'][0]['profile_curves'][4]['end_point']['y'] = p5_y

    data['entities']['entity1']['profiles']['profile1']['loops'][0]['profile_curves'][5]['start_point']['x'] = p5_x
    data['entities']['entity1']['profiles']['profile1']['loops'][0]['profile_curves'][5]['start_point']['y'] = p5_y

    data['entities']['entity1']['profiles']['profile1']['loops'][0]['profile_curves'][5]['end_point']['x'] = p6_x
    data['entities']['entity1']['profiles']['profile1']['loops'][0]['profile_curves'][5]['end_point']['y'] = p6_y

    # Set the center point and radius for the circle
    data['entities']['entity1']['profiles']['profile1']['loops'][1]['profile_curves'][0]['center_point']['x'] = 0.0
    data['entities']['entity1']['profiles']['profile1']['loops'][1]['profile_curves'][0]['center_point']['y'] = 0.0

    data['entities']['entity1']['profiles']['profile1']['loops'][1]['profile_curves'][0]['radius'] = radius

    # Set the area and perimeter properties for the hexagon and circle
    data['entities']['entity1']['profiles']['profile1']['properties']['area'] = outer_area
    data['entities']['entity1']['profiles']['profile1']['properties']['perimeter'] = perimeter_hex

    data['entities']['entity1']['profiles']['profile2']['loops'][0]['profile_curves'][0]['center_point']['x'] = 0.0
    data['entities']['entity1']['profiles']['profile2']['loops'][0]['profile_curves'][0]['center_point']['y'] = 0.0

    data['entities']['entity1']['profiles']['profile2']['loops'][0]['profile_curves'][0]['radius'] = radius

    data['entities']['entity1']['profiles']['profile2']['properties']['area'] = area_circle
    data['entities']['entity1']['profiles']['profile2']['properties']['perimeter'] = perimeter_circle

    # Set the extrusion height
    data['entities']['entity2']['extent_one']['distance']['value'] = height

    return data
```

#### Fill Tube or Ring

```python
def Fill_Tube_Ring(data, shape, height, outer_radius, inner_radius, tube_area=None, tube_perimeter=None, ring_area=None, ring_perimeter=None):
    # Set the component name in the metadata
    data['metadata']['component_name'] = shape

    # Calculate diameters for the outer and inner circles
    d_outer = 2 * outer_radius
    d_inner = 2 * inner_radius

    # Calculate area and perimeter for the inner circle (profile 2)
    area_profile2 = math.pi * inner_radius ** 2
    perimeter_profile2 = 2 * math.pi * inner_radius

    # Update the JSON data with the radii of the outer and inner circles
    data['entities']['entity1']['curves']['outer_circle']['radius'] = outer_radius
    data['entities']['entity1']['curves']['inner_circle']['radius'] = inner_radius

    # Update the diameters in the dimensions section
    data['entities']['entity1']['dimensions']['d_outer']['parameter']['value'] = d_outer
    data['entities']['entity1']['dimensions']['d_inner']['parameter']['value'] = d_inner

    # Update the area in the properties section based on provided values
    if ring_area is not None:
        data['entities']['entity1']['profiles']['profile1']['properties']['area'] = ring_area
    elif tube_area is not None:
        data['entities']['entity1']['profiles']['profile1']['properties']['area'] = tube_area

    # Set the area for the inner circle (profile 2)
    data['entities']['entity1']['profiles']['profile2']['properties']['area'] = area_profile2

    # Update the perimeter in the properties section based on provided values
    if ring_perimeter is not None:
        data['entities']['entity1']['profiles']['profile1']['properties']['perimeter'] = ring_perimeter
    elif tube_perimeter is not None:
        data['entities']['entity1']['profiles']['profile1']['properties']['perimeter'] = tube_perimeter

    # Set the perimeter for the inner circle (profile 2)
    data['entities']['entity1']['profiles']['profile2']['properties']['perimeter'] = perimeter_profile2

    # Update the radii in the loops section for the profiles
    data['entities']['entity1']['profiles']['profile1']['loops'][0]['profile_curves'][0]['radius'] = outer_radius
    data['entities']['entity1']['profiles']['profile1']['loops'][1]['profile_curves'][0]['radius'] = inner_radius
    data['entities']['entity1']['profiles']['profile2']['loops'][0]['profile_curves'][0]['radius'] = inner_radius

    # Set the extrusion height
    data['entities']['entity2']['extent_one']['distance']['value'] = height

    return data
```

### Generate Shape

```python
import json

def Generate_file(shape, height=None, radius=None, side_length=None, length=None, joint_radius=None,
                  nut_area=None, nut_perimeter=None, ring_area=None, ring_perimeter=None,
                  tube_area=None, tube_perimeter=None, bar_area=None, bar_perimeter=None,
                  shackle_area=None, shackle_perimeter=None, outer_radius=None, inner_radius=None):
    """
    Generates a JSON file for a specified shape with given parameters.

    Args:
        shape (str): The shape type (e.g., 'hex nut', 'tube', 'mild square bar', 'ring', 'leaf spring shackle link').
        height (float, optional): The height of the shape.
        radius (float, optional): The radius of the shape.
        side_length (float, optional): The side length of the shape.
        length (float, optional): The length of the shape.
        joint_radius (float, optional): The joint radius for the shackle.
        nut_area (float, optional): The area of the hex nut.
        nut_perimeter (float, optional): The perimeter of the hex nut.
        ring_area (float, optional): The area of the ring.
        ring_perimeter (float, optional): The perimeter of the ring.
        tube_area (float, optional): The area of the tube.
        tube_perimeter (float, optional): The perimeter of the tube.
        bar_area (float, optional): The area of the bar.
        bar_perimeter (float, optional): The perimeter of the bar.
        shackle_area (float, optional): The area of the shackle.
        shackle_perimeter (float, optional): The perimeter of the shackle.
        outer_radius (float, optional): The outer radius of the tube/ring.
        inner_radius (float, optional): The inner radius of the tube/ring.

    Returns:
        str: JSON string representing the shape configuration, or None if the shape is not recognized.
    """

    json_output = None

    if shape == 'hex nut':
        # For hex nut
        json_data = json_hex
        data = json.loads(json_data)
        Hexfile = Fill_Hex(data, shape, height, radius, side_length, nut_area, nut_perimeter)
        json_output = json.dumps(Hexfile, indent=4)

    elif shape == 'tube':
        # For tube
        json_data = json_tube_ring
        data = json.loads(json_data)
        Ring_Tube_file = Fill_Tube_Ring(data, shape, height, outer_radius, inner_radius, tube_area, tube_perimeter)
        json_output = json.dumps(Ring_Tube_file, indent=4)

    elif shape == 'mild square bar':
        # For bar
        json_data = json_bar
        data = json.loads(json_data)
        Bar_file = Fill_Bar(data, shape, height, side_length, bar_area, bar_perimeter)
        json_output = json.dumps(Bar_file, indent=4)

    elif shape == 'ring':
        # For ring
        json_data = json_tube_ring
        data = json.loads(json_data)
        Ring_Tube_file = Fill_Tube_Ring(data, shape, height, outer_radius, inner_radius, ring_area, ring_perimeter)
        json_output = json.dumps(Ring_Tube_file, indent=4)

    elif shape == 'leaf spring shackle link':
        # For shackle
        json_data = json_shackle
        data = json.loads(json_data)
        Shackle_file = Fill_Shackle(data, shape, height, joint_radius, length, shackle_area, shackle_perimeter)
        json_output = json.dumps(Shackle_file, indent=4)

    else:
        print(f'Shape {shape} has no file !!\n')

    return json_output
```

## function

- The Conversion function is designed to generate a Python script that uses the cadquery and gradio libraries to create a 3D model based on the provided JSON data. The generated script initializes a cadquery workplane, processes the JSON data to extract geometric entities, and constructs the corresponding cadquery commands to build the 3D model. Finally, it includes commands to display the model using gradio and export it as an STL file.
- It has user friendly comments

```python
import json

def Conversion(data, name):
    """
    Generates a Python script to create a 3D model using cadquery based on the provided JSON data.

    Args:
        data (str): JSON string containing geometric data.
        name (str): Name of the STL file to be generated.

    Returns:
        str: Python script as a string.
    """
    # Beginning of the code
    commands = "## Uncomment the following lines if you didn't install cadquery and gradio !! \n\n"
    commands += "'''\n!pip install cadquery\n!pip install gradio\n!pip install gradio Pillow\n''' \n\n"
    commands += "## Important Libraries \n\n"
    commands += '''
import pandas as pd
import cadquery as cq
import gradio as gr
from cadquery import exporters
import math
    '''
    commands += "\n# Initialize a new cadquery Workplane object\n"
    commands += "wp = cq.Workplane(\"XY\")\n\n"

    # Reading file
    data = json.loads(data)
    entities = data.get("entities", [])
    points = {}
    curves = {}
    profiles = {}
    extruded_profiles = []

    for entity in data['entities']:
        entity_data = data['entities'][entity]

        if 'type' in entity_data:
            entity_type = entity_data['type']

            if entity_type == 'Sketch':
                # Process points
                if 'points' in entity_data:
                    for point in entity_data['points']:
                        points[point] = entity_data['points'][point]

                # Process curves
                if 'curves' in entity_data:
                    for curve in entity_data['curves']:
                        curve_info = {}
                        curve_type = entity_data['curves'][curve]['type']

                        if curve_type == 'SketchCircle':
                            curve_info = {
                                'type': 'circle',
                                'radius': entity_data['curves'][curve]['radius'],
                                'center': entity_data['curves'][curve]['center_point']
                            }
                        elif curve_type == 'SketchArc':
                            curve_info = {
                                'type': 'arc',
                                'radius': entity_data['curves'][curve]['radius'],
                                'center': entity_data['curves'][curve]['center_point'],
                                'start_angle': entity_data['curves'][curve]['start_angle'],
                                'end_angle': entity_data['curves'][curve]['end_angle'],
                                'start_point': entity_data['curves'][curve]['start_point'],
                                'end_point': entity_data['curves'][curve]['end_point']
                            }
                        elif curve_type == 'SketchLine':
                            curve_info = {
                                'type': 'line',
                                'start_point': entity_data['curves'][curve]['start_point'],
                                'end_point': entity_data['curves'][curve]['end_point']
                            }
                        curves[curve] = curve_info

                # Process profiles
                if 'profiles' in entity_data:
                    for profile in entity_data['profiles']:
                        profile_info = {
                            'Outer_Curves': [],
                            'Inner_Curves': []
                        }
                        for loop in entity_data['profiles'][profile]['loops']:
                            for i in range(len(loop['profile_curves'])):
                                curve_id = loop['profile_curves'][i]['curve']
                                if loop['is_outer']:
                                    profile_info['Outer_Curves'].append(curve_id)
                                else:
                                    profile_info['Inner_Curves'].append(curve_id)
                        profiles[profile] = profile_info

            elif entity_type == 'ExtrudeFeature':
                profile_info = {}
                if 'profiles' in entity_data:
                    for profile in entity_data['profiles']:
                        profile_info['profile'] = profile['profile']
                if 'extent_one' in entity_data:
                    profile_info['distance'] = entity_data['extent_one']['distance']['value']
                    extruded_profiles.append(profile_info)

    # Convert to CADquery commands
    commands += "profile = (\n  wp\n"
    i = 0
    j = 0

    for p in extruded_profiles:
        for c in profiles[p['profile']]['Outer_Curves']:
            curve = curves[c]
            if curve['type'] == 'circle':
                if i != 0 and j != 1:
                    commands += "  .close()\n"
                commands += f"  .moveTo({points[curve['center']]['x']},{points[curve['center']]['y']})\n"
                commands += f"  .circle({curve['radius']})  # Draw Outer Circle\n"
                j = 1
                i += 1
            elif curve['type'] == 'line':
                j = 0
                commands += f"  .moveTo({points[curve['start_point']]['x']},{points[curve['start_point']]['y']})\n"
                commands += f"  .lineTo({points[curve['end_point']]['x']},{points[curve['end_point']]['y']})  # Draw Outer Line\n"
                if i == len(profiles[p['profile']]['Outer_Curves']):
                    commands += "  .close()\n"
                i += 1
            elif curve['type'] == 'arc':
                commands += f"  .moveTo({points[curve['start_point']]['x']},{points[curve['start_point']]['y']})\n"
                radius = curve['radius']
                if data['metadata']['component_name'] == 'leaf spring shackle link':
                    radius = -1 * (radius + 0.001)
                commands += f"  .radiusArc(({points[curve['end_point']]['x']},{points[curve['end_point']]['y']}), {radius})  # Draw Outer Arc\n"
                if i == len(profiles[p['profile']]['Outer_Curves']):
                    commands += "  .close()\n"
                i += 1

    i = 0
    j = 0

    for p in extruded_profiles:
        for c in profiles[p['profile']]['Inner_Curves']:
            curve = curves[c]
            if curve['type'] == 'circle':
                if i != 0 and j != 1:
                    commands += "  .close()\n"
                commands += f"  .moveTo({points[curve['center']]['x']},{points[curve['center']]['y']})\n"
                commands += f"  .circle({curve['radius']})  # Draw Inner Circle\n"
                j = 1
                i += 1
            elif curve['type'] == 'line':
                j = 0
                commands += f"  .moveTo({points[curve['start_point']]['x']},{points[curve['start_point']]['y']})\n"
                commands += f"  .lineTo({points[curve['end_point']]['x']},{points[curve['end_point']]['y']})  # Draw Inner Line\n"
                if i == len(profiles[p['profile']]['Inner_Curves']):
                    commands += "  .close()\n"
                i += 1
            elif curve['type'] == 'arc':
                commands += f"  .moveTo({points[curve['start_point']]['x']},{points[curve['start_point']]['y']})\n"
                radius = curve['radius']
                if data['metadata']['component_name'] == 'leaf spring shackle link':
                    radius = -1 * (radius + 0.001)
                commands += f"  .radiusArc(({points[curve['end_point']]['x']},{points[curve['end_point']]['y']}), {radius})  # Draw Inner Arc\n"
                if i == len(profiles[p['profile']]['Inner_Curves']):
                    commands += "  .close()\n"
                i += 1

    commands += ")\n"

    # Extrude
    for p in extruded_profiles:
        commands += f"profile = profile.extrude({p['distance']})\n"

    # Gradio
    commands += "\n\n# For Gradio\n"
    commands += '''
# Export the result to a file in SVG format.
exporters.export(profile, "C:\\\\Users\\\\tibra\\\\shape.svg")

def display_svg():
    with open("C:\\\\Users\\\\tibra\\\\shape.svg", "r") as svg_file:
        svg_content = svg_file.read()
    return svg_content

iface = gr.Interface(
    fn=display_svg,
    inputs=None,
    outputs="html",
    title="SVG Image Display",
    description="Display an SVG image from the specified path."
)

iface.launch()
    '''

    # Drive
    commands += '''
# To mount to your Drive
from google.colab import drive
drive.mount('/content/drive')
    '''

    # STL
    commands += '''
# Export the result as an STL file to the specified path
exporters.export(profile, f'/content/drive/MyDrive/website/STL/{name}.stl')
    '''

    return commands
```
