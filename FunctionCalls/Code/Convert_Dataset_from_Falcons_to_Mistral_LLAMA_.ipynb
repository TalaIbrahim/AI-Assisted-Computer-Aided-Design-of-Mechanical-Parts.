Converting the dataset from Falcons #user_promt #Assistant Format to Mistral/LLama \<S>\<INST> Format

```python
# Import necessary libraries

from google.colab import drive

import json
import pandas as pd
```

```python
# Mount Google Drive to access files

drive.mount('/content/drive/')

# Define the word to remove from the JSON content

word_to_remove = 'text:'
```

```python
# Read the JSON file from Google Drive

with open('/content/drive/MyDrive/PROJECT2/function_call/FunctionCALLfalcondata.json', 'r') as file:
    content = ''
    content = file.read()





# Perform replacements on the content
modified_content = content.replace('{"text": ', '')

modified_content = modified_content.replace('### user_promt:', "<s>[INST]").replace("### Assistant:", "[/INST]").replace('"}','</s>"')
```

```python
# Save the modified content to a new file

output_filename = f'/content/drive/MyDrive/PROJECT2/function_call/FunctionCALLmistraldata.txt'
with open(output_filename, 'w') as output_file:
    output_file.write(modified_content)
```
